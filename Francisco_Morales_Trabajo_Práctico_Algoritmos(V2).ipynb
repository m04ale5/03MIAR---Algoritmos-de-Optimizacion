Espacio de soluciones: El espacio de soluciones del problema de programación de partidos de fútbol es potencialmente grande y está determinado por la cantidad de combinaciones de horarios disponibles para cada jornada, la asignación de equipos a esos horarios, y el cumplimiento de las restricciones impuestas.

Técnica Utilizada: Opté por utilizar el algoritmo genético, ya que es una técnica de optimización que imita la evolución biológica, empleando operadores genéticos como selección, cruce y mutación para buscar soluciones óptimas en un espacio de búsqueda.
Razón de Uso: Consideré que el AG sería adecuado para abordar el problema de asignación de horarios de partidos de fútbol, ya que maneja eficientemente un amplio espacio de búsqueda y puede encontrar soluciones aceptables en un tiempo razonable.<br>

## Los cálculos asociados a una jornada de ejemplo

Los cálculos asociados a una jornada de ejemplo se realizaron con el propósito de verificar la validez del método bajo las restricciones establecidas, con el fin de comparar los resultados obtenidos con el ejemplo presentado en clase.
```
import random

# Definir las categorías de equipos y la cantidad de equipos en cada categoría
categorias_equipos = {
    'A': 3,
    'B': 11,
    'C': 6
}

# Definir los horarios disponibles y sus coeficientes de audiencia
horarios_disponibles = {
    'Viernes': {'20': 0.4},
    'Sábado': {'12': 0.55, '16': 0.7, '18': 0.8, '20': 1},
    'Domingo': {'12': 0.45, '16': 0.75, '18': 0.85, '20': 1},
    'Lunes': {'20': 0.4}
}

# Definir la audiencia de cada tipo de partido
audiencia_partidos = {
    ('A', 'A'): 2,
    ('A', 'B'): 1.3,
    ('A', 'C'): 1,
    ('B', 'B'): 0.9,
    ('B', 'C'): 0.75,
    ('C', 'C'): 0.47,
    ('B', 'A'): 1.3,
    ('C', 'A'): 1,
    ('C', 'B'): 0.75
}

# Tabla de reducción de audiencia según el número de coincidencias de horarios
reduccion_coincidencias = {
    1: 0.0,
    2: 0.25,
    3: 0.45,
    4: 0.6,
    5: 0.7,
    6: 0.75,
    7: 0.78,
    8: 0.8
}

# Generar un horario aleatorio para un partido
def calcular_audiencia_total_fija(jornada, horarios_fijos, horarios_disponibles):
    audiencia_total = 0
    coincidencias_horarios = {}

    # Contabilizar el número de coincidencias para cada horario
    for partido, (dia, hora) in horarios_fijos.items():
        if (dia, hora) in coincidencias_horarios:
            coincidencias_horarios[(dia, hora)] += 1
        else:
            coincidencias_horarios[(dia, hora)] = 1

    # Calcular la audiencia total teniendo en cuenta las coincidencias
    for partido, equipos in jornada.items():
        dia, hora = horarios_fijos[partido]  # Obtener el horario fijo del partido
        coeficiente_audiencia = horarios_disponibles[dia][hora]
        audiencia_partido = audiencia_partidos[equipos]  # Usar la categoría de los equipos para buscar la audiencia del partido

        # Obtener el número de coincidencias para el horario actual
        num_coincidencias = coincidencias_horarios.get((dia, hora), 0)

        # Si no hay coincidencias, la reducción es nula (0)
        if num_coincidencias == 0:
            reduccion = 0
        else:
            # Calcular el coeficiente de reducción basado en el número de coincidencias
            reduccion = reduccion_coincidencias.get(num_coincidencias, 0)

        # Calcular la audiencia neta del partido con la reducción de audiencia
        audiencia_partido_neta = audiencia_partido * coeficiente_audiencia * (1 - reduccion)
        audiencia_total += audiencia_partido_neta

        # Mostrar los detalles del cálculo del partido
        print(f"Partido: {partido}, Equipo: {equipos}, Día: {dia}, Hora: {hora}, Coeficiente de Audiencia: {coeficiente_audiencia}, Audiencia del Partido: {audiencia_partido}, Reducción: {reduccion}, Audiencia Neta: {audiencia_partido_neta}")

    return audiencia_total


if __name__ == "__main__":
    jornada = [('B', 'A'), ('B', 'A'), ('C', 'C'), ('B', 'A'), ('C', 'C'), ('B', 'C'), ('B', 'B'), ('B', 'B'), ('B', 'C'), ('B', 'B')]

    # Convertir la lista de tuplas a un diccionario con números de partido como claves
    jornada_dict = {i: partido for i, partido in enumerate(jornada)}

    # Definir los horarios fijos
    horarios_fijos = {
        0: ('Viernes', '20'),
        1: ('Sábado', '12'),
        2: ('Sábado', '16'),
        3: ('Sábado', '18'),
        4: ('Sábado', '20'),
        5: ('Domingo', '16'),
        6: ('Domingo', '16'),
        7: ('Domingo', '18'),
        8: ('Domingo', '20'),
        9: ('Lunes', '20')
    }

    # Calcular la audiencia total con los horarios fijos
    audiencia_total_jornada_fija = calcular_audiencia_total_fija(jornada_dict, horarios_fijos, horarios_disponibles)
    print("Audiencia Total con Horarios Fijos:", audiencia_total_jornada_fija)
```

## Organizar los horarios de partidos de La Liga con Algoritmo Genético

```
import random

# Categorías de equipos con la cantidad de equipos en cada categoría
categorias_equipos = {
    'A': 3,
    'B': 11,
    'C': 6
}

# Horarios disponibles con coeficientes de audiencia para cada franja horaria
horarios_disponibles = {
    'Viernes': {'20': 0.4},
    'Sábado': {'12': 0.55, '16': 0.7, '18': 0.8, '20': 1},
    'Domingo': {'12': 0.45, '16': 0.75, '18': 0.85, '20': 1},
    'Lunes': {'20': 0.4}
}

# Audiencia esperada para diferentes combinaciones de equipos
audiencia_partidos = {
    ('A', 'A'): 2,
    ('A', 'B'): 1.3,
    ('A', 'C'): 1,
    ('B', 'B'): 0.9,
    ('B', 'C'): 0.75,
    ('C', 'C'): 0.47,
    ('B', 'A'): 1.3,
    ('C', 'A'): 1,
    ('C', 'B'): 0.75
}

# Reducción en la audiencia debido a coincidencias de horarios
reduccion_coincidencias = {
    1: 0.0,
    2: 0.25,
    3: 0.45,
    4: 0.6,
    5: 0.7,
    6: 0.75,
    7: 0.78,
    8: 0.8
}

# Genera un horario aleatorio (día y hora)
def generar_horario_aleatorio():
    dia = random.choice(list(horarios_disponibles.keys()))
    hora = random.choice(list(horarios_disponibles[dia].keys()))
    return dia, hora

# Calcula la audiencia total para una jornada y una asignación de horarios
def calcular_audiencia_total(jornada, horarios_fijos):
    audiencia_total = 0
    coincidencias_horarios = {}

    # Contar las coincidencias de horarios
    for partido, (dia, hora) in horarios_fijos.items():
        if (dia, hora) in coincidencias_horarios:
            coincidencias_horarios[(dia, hora)] += 1
        else:
            coincidencias_horarios[(dia, hora)] = 1

    # Calcular la audiencia total considerando los horarios asignados
    for partido, equipos in jornada.items():
        dia, hora = horarios_fijos[partido]
        coeficiente_audiencia = horarios_disponibles[dia][hora]
        audiencia_partido = audiencia_partidos[equipos]

        num_coincidencias = coincidencias_horarios.get((dia, hora), 0)

        # Aplicar reducción de audiencia por coincidencias de horarios
        if num_coincidencias == 0:
            reduccion = 0
        else:
            reduccion = reduccion_coincidencias.get(num_coincidencias, 0)

        audiencia_partido_neta = audiencia_partido * coeficiente_audiencia * (1 - reduccion)
        audiencia_total += audiencia_partido_neta

    return audiencia_total

# Genera una población inicial de horarios aleatorios
def poblacion_inicial(tamano_poblacion, jornada):
    poblacion = []
    for _ in range(tamano_poblacion):
        solucion = {}
        for index, partido in enumerate(jornada):
            solucion[index] = generar_horario_aleatorio()
        poblacion.append(solucion)
    return poblacion

# Algoritmo genético para optimizar los horarios de la jornada
def algoritmo_genetico(jornada, tamano_poblacion, probabilidad_cruce, probabilidad_mutacion, num_generaciones, elitismo=0.2, probabilidad_mutacion_individuo=0.):
    poblacion = poblacion_inicial(tamano_poblacion, jornada)

    for _ in range(num_generaciones):
        poblacion = sorted(poblacion, key=lambda x: calcular_audiencia_total(jornada, x), reverse=True)

        num_elitismo = int(elitismo * tamano_poblacion)
        mejores_soluciones = poblacion[:num_elitismo]

        poblacion_con_busqueda_local = []

        padres_cruce = poblacion[:tamano_poblacion // 2 - num_elitismo]
        descendientes_cruce = []

        # Cruzar padres para generar descendientes
        while len(descendientes_cruce) < tamano_poblacion - num_elitismo:
            padre1, padre2 = random.choices(padres_cruce, k=2)
            hijo1, hijo2 = cruce(padre1, padre2)
            descendientes_cruce.extend([hijo1, hijo2])

        poblacion_mutacion = random.sample(poblacion, int(probabilidad_mutacion_individuo * tamano_poblacion))
        descendientes_mutacion = [mutacion(individuo, probabilidad_mutacion) for individuo in poblacion_mutacion]

        poblacion = poblacion_con_busqueda_local + descendientes_cruce + descendientes_mutacion

    mejor_solucion = max(poblacion, key=lambda x: calcular_audiencia_total(jornada, x))
    mejor_audiencia = calcular_audiencia_total(jornada, mejor_solucion)

    return mejor_solucion, mejor_audiencia

# Realiza el cruce de dos soluciones para generar dos descendientes
def cruce(padre1, padre2):
    punto_cruce = random.randint(0, len(padre1) - 1)
    hijo1 = {k: padre1[k] if k < punto_cruce else padre2[k] for k in range(len(padre1))}
    hijo2 = {k: padre2[k] if k < punto_cruce else padre1[k] for k in range(len(padre1))}
    return hijo1, hijo2

# Realiza la mutación en un individuo de la población
def mutacion(individuo, probabilidad_mutacion):
    for key in individuo.keys():
        if random.random() < probabilidad_mutacion:
            individuo[key] = generar_horario_aleatorio()
    return individuo

if __name__ == "__main__":
    # Definición de la jornada de partidos
    jornada = {i: partido for i, partido in enumerate([('B', 'A'), ('B', 'A'), ('C', 'C'), ('B', 'A'), ('C', 'C'), ('B', 'C'), ('B', 'B'), ('B', 'B'), ('B', 'C'), ('B', 'B')])}
    tamano_poblacion = 2000
    probabilidad_cruce = 0.95
    probabilidad_mutacion = 0.10
    num_generaciones = 500

    # Solución inicial aleatoria
    solucion_aleatoria = {i: generar_horario_aleatorio() for i in range(len(jornada))}

    # Calcular audiencia inicial
    audiencia_total_jornada = calcular_audiencia_total(jornada, solucion_aleatoria)
    print("Audiencia Inicial total de la jornada:", audiencia_total_jornada)

    # Ejecutar el algoritmo genético para optimizar los horarios
    mejor_solucion, mejor_audiencia = algoritmo_genetico(jornada, tamano_poblacion, probabilidad_cruce, probabilidad_mutacion, num_generaciones)

    # Imprimir la mejor combinación de partidos y horarios encontrada
    print("Mejor combinación de partidos y horarios:")
    for partido, horario in mejor_solucion.items():
        print(f"Partido {partido + 1}: {horario[0]} a las {horario[1]}")
    print(f"Audiencia total: {mejor_audiencia:.2f}")
```

Tras aplicar el algoritmo genético, obtuve una audiencia inicial total de la jornada de 4.48. Después de la optimización, logré encontrar una combinación de partidos y horarios que maximizó la audiencia. La mejor distribución resultó en una audiencia total de 6.86, lo que representa un incremento significativo con respecto a la audiencia inicial.

La estrategia del algoritmo genético fue efectiva para encontrar una programación óptima de los partidos, distribuyéndolos de manera más favorable a lo largo del fin de semana.
